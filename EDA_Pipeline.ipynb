### Remove elements by index

You can remove elements from a Series by specifying their index using the `drop` method.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
df = pd.read_csv("cars_data.csv")
```

```python
display(df.head())
```

```python
display(df.tail())
```

```python
df.dtypes
```

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

```python
df = df.rename(columns = {"Engine HP": "HP", "Engine Cylinders": "Total Cylinders", "Transmission Type": "Mechanism", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

```python
df.shape
```

```python
df.count()
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df = df.dropna()
df.count()
```

```python
sns.boxplot(x=df['Price'])
plt.show()
```

```python
sns.boxplot(x=df['HP'])
plt.show()
```

```python
sns.boxplot(x=df['Total Cylinders'])
plt.show()
```

```python
numeric_cols = df.select_dtypes(include='number').columns
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

conditions = ~((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)

df_cleaned = df[conditions]

print(df_cleaned.shape)
```
